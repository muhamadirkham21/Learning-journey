```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

##  Introduction

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
```

### inisiasi figure dan axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y)
```

```python
y1 = np.sin(X)
y2 = np.cos(X)

fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1, label ='sin')
ax.plot(X, y2, label ='cos')
ax.set(title='grafik contoh', xlabel='apasih')
ax.legend();
```

### shortcut

```python
plt.figure(figsize=(4, 4))
plt.plot(X, y1, 'r--', linewidth=0.8)
plt.ylim(-2, 2)
plt.plot(X, y2)
```

# common plot

```python
# membuat barplot
# barplot digunakan untuk visualisasi univariate

 
cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta', 
          'Surakarta','Surabaya', 'Medan', 'Makassar')
 
populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)
 
plt.bar(x=cities, height=populations)
plt.xticks(rotation=40)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    'cities' : cities,
    'populations' : populations
})

df.sort_values(by='populations', inplace=True)

plt.barh(y=df['cities'], width=df['populations'])
plt.xlabel('populations(millions)')
plt.title('populations in indonesia')
plt.show()
```

```python
# pie chart

flavors = ('Chocolate', 'Vanilla', 'Macha', 'Others')
votes = (50, 20, 30, 10)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.1, 0, 0, 0)
 
plt.pie(
    x=votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)
plt.show()
```

```python
# donut chart
plt.pie(
    x=votes,
    labels=flavors,
    colors=colors,
    wedgeprops = {'width': 0.4}
)
plt.show()
```

```python
# histogram
import numpy as np

x = np.random.normal(15, 5, 250)
 
plt.hist(x=x, bins=15)
plt.show()
```

```python
# visualisasi bivariate
# scatter plot

lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15, 
                  9.96, 7.21, 10.04, 10.2, 11.06]
lemon_weight = [112.05, 114.58, 116.71, 117.4, 128.93, 
                132.93, 138.92, 145.98, 148.44, 152.81]

plt.scatter(x=lemon_diameter, y=lemon_weight)
plt.show()
```

```python
sns.scatterplot(x=lemon_diameter, y=lemon_weight)
plt.show()
```

```python
sns.regplot(x=lemon_diameter, y=lemon_weight)
plt.show()
```

```python
# line chart
url = 'https://query1.finance.yahoo.com/v7/finance/download/BBCA.JK?period1=1644796800&period2=1676332800&interval=1d&events=history&includeAdjustedClose=true'
df = pd.read_csv(url)
df['Date'] = pd.to_datetime(df['Date'])
 
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Close'], color='red')
plt.xlabel('Date',size=15)
plt.ylabel('Price',size=15)
plt.show()
```

```python
# bar chart per kelompok

penguins = sns.load_dataset("penguins")
 
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", ci=None)
plt.show()
```

```python
# scatter plot berkelompok
sns.scatterplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="species", style="species")
plt.show()
```

# 3d plot

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x, y)
z = np.sqrt(X**2 + Y**2 + 10)

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z)
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x, y)
z = np.sqrt(X**2 + Y**2 + 10)

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, antialiased=False, cmap='winter')
```
