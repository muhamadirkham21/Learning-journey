```python
def read_stabil_video(path):

    capture = cv.VideoCapture(path)

    if not capture.isOpened():
        print("Error: Could not open video file.")
        exit()
    
    while True:
        ret, frame = capture.read()

        if not ret:
            print("End of video")
            break

        cv.imshow('Video', frame)

        if cv.waitKey(25) & 0xFF == ord('q'):
            break


    capture.release()
    cv.destroyAllWindows()

read_stabil_video('/home/irkham/Learning-journey/learn-python/ADVANCED/COMPUTER VISION/videos/video_1.mp4')
```

## membuat blank image

```python
import numpy as np

blank = np.zeros((500, 500), dtype='uint8')
cv.imshow('blank', blank)

cv.waitKey(0)
cv.destroyAllWindows()
```

## blank berwarna


```python
import numpy as np

blank = np.zeros((500, 500, 3), dtype='uint8')
blank[:] = 0, 255, 0
cv.imshow('blank', blank)

cv.waitKey(0)
cv.destroyAllWindows()
```

## membuat persegi

```python
import cv2 as cv
import numpy as np 

blank = np.zeros((500, 500, 3), dtype='uint8')
cv.rectangle(blank, (0, 0), (250, 250), (0, 0, 255), thickness=2)
cv.imshow('rect', blank)

cv.waitKey(0)
cv.destroyAllWindows()
```

## membuat text

```python
blank = np.zeros((500, 500, 3), dtype='uint8')
cv.putText(blank, 'hello', (255, 255), cv.FONT_HERSHEY_COMPLEX, 1.0, (255, 0, 0), 2)
cv.imshow('text', blank)

cv.waitKey(0)
cv.destroyAllWindows()
```

## convert to grey scale image

```python
img = cv.imread('/home/irkham/Learning-journey/learn-python/ADVANCED/COMPUTER VISION/images/cat_2.jpg')

grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Grey', grey)

cv.waitKey(0)
cv.destroyAllWindows()
```

## blur image

```python
img = cv.imread('/home/irkham/Learning-journey/learn-python/ADVANCED/COMPUTER VISION/images/cat_2.jpg')

blur = cv.GaussianBlur(img, (7, 7), cv.BORDER_DEFAULT)

cv.imshow('blur', blur)

cv.waitKey(0)
cv.destroyAllWindows()
```

## melihat fitur gambar (edge cascade)
