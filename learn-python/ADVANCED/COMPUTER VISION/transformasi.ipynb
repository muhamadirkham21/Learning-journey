```python
import cv2 as cv
import numpy as np
```

```python
def translate(img, x, y):
    transMat = np.float32([[1, 0, x],[0, 1, y]])
    dimension = (img.shape[1], img.shape[0])

    return cv.warpAffine(img, transMat, dimension)

img = cv.imread('/home/irkham/Learning-journey/learn-python/ADVANCED/COMPUTER VISION/images/cat_2.jpg')
cv.imshow('img', img)

trans_right = translate(img, 100, 100)
cv.imshow('trans', trans_right)


cv.waitKey(0)
cv.destroyAllWindows()
```

## rotation

```python
def rotate(img, angle, rotPoint=None):
    (height, width) = img.shape[:2]

    if rotate is None:
        rotPoint = (width//2, height//2)
    
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1.0)
    dimension = (width, height)
    return cv.warpAffine(img, rotMat, dimension)


img = cv.imread('/home/irkham/Learning-journey/learn-python/ADVANCED/COMPUTER VISION/images/cat_2.jpg')
cv.imshow('img', img)

rotate = rotate(img, -45)
cv.imshow('rotate', rotate)


cv.waitKey(0)
cv.destroyAllWindows()
    
```
