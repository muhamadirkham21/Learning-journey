```python
import cv2 as cv
import numpy as np
```

```python
img = cv.imread('/home/irkham/Learning-journey/learn-python/ADVANCED/COMPUTER VISION/images/cat_2.jpg')
cv.imshow('image', img)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('greyscale', gray)

blur = cv.GaussianBlur(gray, (5, 5), cv.BORDER_DEFAULT)
cv.imshow('blur', blur)

canny = cv.Canny(blur, 125, 175)
cv.imshow('canny', canny)

cv.waitKey(0)
cv.destroyAllWindows()
```
